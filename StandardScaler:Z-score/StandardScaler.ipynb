```python
import pandas as pd
from sklearn.preprocessing import StandardScaler # import StandardScaler from the sklearn library
import matplotlib.pyplot as plt # impporting the matplotlib.pyplot to show graph

%matplotlib inline 
# matplotlib inline is used to show inline graph
scaler = StandardScaler() # modify StandardScaler as scaler

data = pd.read_csv('diabetes_DP_StandardScaler.csv') # open csv file

X = data.drop(['preg','class'], axis = 1) # get all col names from the table except 'preg' and 'class' col
Y = data['preg'] # get only 'preg' col of the table
 
scaler.fit_transform(X) # use 'z = (x - u) / s' formula is used to each table row and change all table values

plt.plot(X,Y) # to illustrate in graph



#data.head()  show the first 5 elements of the table
#data.tail()  show the last 5 elements of the table
# print(data.isna().any()) to check wheter data is clean or not
```
