```python
import numpy as np
import pandas as pd
from numpy.random import randn 
from matplotlib import pyplot as plt

%matplotlib inline

X = [12, 15, 17, 18, 20, 24, 27, 31, 40, 45]

Y = [220, 231, 324, 355, 380, 391, 400, 422, 433, 500]

Z = [121, 130, 145, 165, 180, 200, 240, 324, 457, 500]

plt.plot(X,Y) # show line graph
plt.plot(X,Z) # show line graph

plt.xlabel('Age') # show label X axis
plt.ylabel('Costs') # show label to Y label
plt.title('Age to Cost') # add title

plt.show() # show graph labels
```

```python
#HISTOGRAM

# create a DataFrame
data = pd.DataFrame(randn(50,4), columns = ['January', 'February', 'March', 'April'])

data['March'].hist(bins = 50, grid = False) # bins shows the widths of hist and grid takes of lines 
```

```python
#AREA

data.plot(kind = 'area' , stacked = False, alpha = 0.5, title = 'Area Plot Exaample', legend = False)
# stacked : used to solve negative number and positive numbers , alpha is opecity , legend takes legend elements
```

```python
# SCATTER

data.plot(kind = 'scatter', x= 'January', y = 'March', c = 'February', cmap = 'magma')
# cmap = 'magma' is the type of color mostly used
```

```python
# BOX

data.plot(kind ='box' , figsize = (10, 5), cmap = 'magma', grid = True, title = 'Box Plot' )
#figsize just size of boxes 
```

```python
#DENSITY 
# density and kde are the same graph I can use
data.plot(kind = 'density', title = 'denisty graph', grid = True)
```

```python
# HEXBIN

data.plot(kind = 'hexbin', x='January', y = 'February', gridsize = 10)
```

```python
# Dansity and KDE
# if I don't have kde, I can use 'pip install scipy', lw='' use to the with of the lines,  alpha is obcity
data.plot(kind = 'kde', lw = 4, alpha = .5)
```

```python
#BAR CHART

#.abs() can use to avoid from negative values, staked use to add all values and make one column 
data.abs().plot(kind = 'bar', stacked = True) 
```
